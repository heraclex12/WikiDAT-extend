```python
from elasticsearch import Elasticsearch
import pandas as pd
import csv
import json
import re
import os
```

```python
es = Elasticsearch(['localhost'], timeout=60)
es.indices.exists('viwiki_history')
```

```python
if not os.path.exists('data'):
    os.mkdir('data')
    print("Created data/ folder")
```

```python
def get_revs(offset):
    res = es.search(index='viwiki_history', body={
            'query':{
                'match':{
                    "comment":"lỗi chính tả"
                }
            }
        }, size=10000, from_ = offset)
    
    return res
```

```python
def write_to_file(rev):
    parent_id = rev['_source']['parent_id']
    current_id = rev['_id']
    folder_name = 'data/' + str(current_id) + '_' + str(parent_id)
    os.mkdir(folder_name)
    try:
        mis_rev = es.get(index='viwiki_history', id=parent_id)

        with open(folder_name + '/info.txt', 'w') as f:
            info = {
                'current_id' : current_id,
                'parent_id' : parent_id,
                'page_id' : rev['_source']['page_id']
            }
            json.dump(info, f, indent=4)

        with open(folder_name + '/good_content.txt', 'w', encoding='utf-8') as f:
            f.write("\n".join(rev['_source']['content']))


        with open(folder_name + '/bad_content.txt', 'w', encoding='utf-8') as f:
            f.write("\n".join(mis_rev['_source']['content']))
    
        print("Save content -> OK")
    
    except:
        print("Not found:", parent_id, current_id)
```

```python
res = get_revs(0)
```

```python
cnt = 0
total_content = 0
rev_ids = set()
duplicate_ids = set()


for rev in res['hits']['hits']:
    cnt += 1
    if cnt % 500 == 0:
        print("Cross:", cnt)
        
    if rev['_source']['page_id'] not in rev_ids:
        rev_ids.add(rev['_id'])
        if len("\n".join(rev['_source']['content'])) > 200:
            write_to_file(rev)
            total_content += 1

    else:
        duplicate_ids.add(rev['_id'])
        print('Duplicate revision: ', rev['_id'], rev['_source']['parent_id'])
```
